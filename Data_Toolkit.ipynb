*THEORY*

- 1. What is NumPy, and why is it widely used in Python?
    - NumPy is a Python library used for working with arrays. It is widely used because it makes numerical calculations much faster and easier compared to normal Python lists. It also supports many mathematical functions.

- 2. How does broadcasting work in NumPy?
    - Broadcasting helps NumPy perform operations on arrays of different sizes. Instead of giving an error, NumPy automatically adjusts the smaller array so both arrays can be used together in calculations.

- 3. What is a Pandas DataFrame?
    - A DataFrame is a table-like structure in Pandas where data is arranged in rows and columns. It is easy to work with and helps in cleaning and analyzing data.

- 4. Explain the use of the groupby() method in Pandas.
    - groupby() is used to group the data based on a column and then perform operations like sum, mean, count, etc., on each group separately.

- 5. Why is Seaborn preferred for statistical visualizations?
    - Seaborn is preferred because it gives better-looking graphs and makes it easy to create statistical plots with very simple code.

- 6. What are the differences between NumPy arrays and Python lists?
    - NumPy arrays are faster, take less memory, and allow mathematical operations directly. Python lists are slower and do not support math operations without loops.

- 7. What is a heatmap, and when should it be used?
    - A heatmap is a graph where values are shown using colors. It is mainly used to show patterns or relationships, like correlation between variables.

- 8. What does the term “vectorized operation” mean in NumPy?
    - Vectorized operations mean performing calculations on entire arrays at once instead of using loops. This makes programs run faster.

- 9. How does Matplotlib differ from Plotly?
    - Matplotlib creates simple and static graphs, while Plotly creates interactive graphs where we can zoom, hover, and explore data easily.

- 10. What is the significance of hierarchical indexing in Pandas?
    - Hierarchical indexing lets us use more than one index level. It helps handle complex datasets in a cleaner way.

- 11. What is the role of Seaborn’s pairplot() function?
    - pairplot() shows scatter plots and distributions for all numeric columns, helping us quickly understand relationships between variables.

- 12. What is the purpose of the describe() function in Pandas?
    - describe() gives basic statistics of the dataset like mean, max, min, count, and percentiles.

- 13. Why is handling missing data important in Pandas?
    - Missing data can affect the accuracy of results, so it is important to clean or fill the missing values before analysis.

- 14. What are the benefits of using Plotly for data visualization?
    - Plotly helps create interactive and attractive charts. It is user-friendly and works well for dashboards or web-based visualizations.

- 15. How does NumPy handle multidimensional arrays?
    - NumPy stores multidimensional arrays efficiently in memory and allows us to access elements using indexes for each dimension.

- 16. What is the role of Bokeh in data visualization?
    - Bokeh is used to create interactive charts that run in a browser and are useful for building dashboards.

- 17. Explain the difference between apply() and map() in Pandas.
    - map() is used on Series for simple element-wise operations. apply() can be used on both Series and DataFrames and can run more complex functions.

- 18. What are some advanced features of NumPy?
    - Some advanced features include broadcasting, linear algebra operations, random number generation, and working with large multi-dimensional arrays.

- 19. How does Pandas simplify time series analysis?
    - Pandas provides tools for date-time indexing, resampling, shifting, and rolling calculations, which makes time-related analysis easy.

- 20. What is the role of a pivot table in Pandas?
    - A pivot table helps us summarize data by rearranging columns and applying functions like sum or average.

- 21. Why is NumPy’s array slicing faster than Python’s list slicing?
    - NumPy slicing is faster because it creates a view of the same data without copying it. Python list slicing creates a new list, which takes more time.

- 22. What are some common use cases for Seaborn?
    - Seaborn is commonly used for heatmaps, pairplots, distribution plots, and categorical plots like bar plots and box plots.

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

df = pd.DataFrame({"marks": [50, 60, 70, 80]})

mean_value = df["marks"].mean()
print(mean_value)
```

```python
# 3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [5, 3, 6, 7]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# 4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap.
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [4, 5, 6, 7],
    "C": [7, 8, 9, 10]
})

corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# 5. Generate a bar plot using Plotly.
import plotly.express as px

data = {"names": ["A", "B", "C"], "values": [10, 20, 30]}

fig = px.bar(data, x="names", y="values")
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd

df = pd.DataFrame({"price": [100, 200, 300]})

df["discount"] = df["price"] * 0.9
print(df)
```

```python
# 7. Perform element-wise multiplication of two NumPy arrays.
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [2, 3, 4, 5]
y2 = [5, 4, 3, 2]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

df = pd.DataFrame({"age": [12, 18, 25, 30]})

filtered = df[df["age"] > 20]
print(filtered)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 15, 30, 25, 40]

sns.histplot(data, bins=5, kde=True)
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy.
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = np.dot(A, B)
print(result)
```
